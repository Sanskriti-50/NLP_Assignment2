```python
train_set, validation_set = train_test_split(proc_sents, test_size=0.2)
```

```python
word_to_int = {}
count = 0
for sentence in train_set:
    for token in sentence:
        if(token not in word_to_int): 
            word_to_int[token] = count
            count += 1
```

```python
unigrams = {}
bigrams = {}
trigrams = {}
quadgrams = {}
```

```python
for sentence in train_set:
    for token in sentence:
        if(token not in unigrams): unigrams[token] = 0
        unigrams[token] += 1
```

```python
total_count0 = 0
for value in unigrams.values():
    total_count0 += value 
    
total_count0
```

```python
total_perp = 0
count_sent = 0

for sentence in validation_set:
    
    count_sent += 1
    if(len(sentence) <=0): continue
    else:
        perp = 1
        n = len(sentence)
        log_perp=0
        for i in range(0 , n):
            n_gram = (sentence[i])
            
            if(n_gram not in unigrams):
                log_perp = math.log(10**7)#Make fla('inf')
                break
                
            else:
                log_perp += math.log(1/(unigrams[n_gram]/total_count0))
#         print(log_perp)
        total_perp += math.exp((1/n)*log_perp)
print("average perplexity of unigram is: ",total_perp / count_sent)
```

## Laplace Smoothing 

we will add 1 in numerator and V which is number of unique words to denominator for all the probabilities

```python
total_perp = 0
count_sent = 0

for sentence in validation_set:
    
    count_sent += 1
    if(len(sentence) <1): continue
    else:
        perp = 1
        n = len(sentence)
        log_perp=0
        for i in range(0 , n):
            n_gram = (sentence[i])
            
            if(n_gram not in unigrams):
                log_perp = math.log(1.0 / (total_count0 + v))
                break
            else:
                
            
                log_perp += math.log(1/(unigrams[n_gram]+1/total_count0+v))
#         print(log_perp)
        total_perp += math.exp((1/n)*log_perp)
print("average perplexity of unigram is: ",total_perp / count_sent)
```

```python
total_perp = 0
count_sent = 0

for sentence in validation_set:
    
    count_sent += 1
    if(len(sentence) < 1): continue
    else:
        perp = 1
        n = len(sentence)
        log_perp=0
        for i in range(1 , n):
            n_gram = (sentence[i-1], sentence[i])
            
            if(n_gram not in bigrams):
                if n_gram[0] in unigrams:
                    log_perp = math.log(1.0 / (unigrams[n_gram[0]] + v))
                else:
                    log_perp = math.log(1.0 / v)  
                break
                
            else:
                log_perp += math.log(1/((bigrams[n_gram]+1)/unigrams[n_gram[0]]+v))
        total_perp += math.exp((1/n)*log_perp)
print(" after Laplace smoothing average perplexity of bigram is: ",total_perp / count_sent)
```
